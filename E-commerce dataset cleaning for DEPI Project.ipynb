```python
import pandas as pd
import numpy as np

```

```python
try:
    df_customers = pd.read_csv("customers.csv")
    df_orders = pd.read_csv("orders.csv")
    df_order_items = pd.read_csv("order_items.csv")
    df_payments = pd.read_csv("payments.csv")
    df_products = pd.read_csv("products.csv")

    print("All 5 files loaded successfully:")
    print(" - df_customers")
    print(" - df_orders")
    print(" - df_order_items")
    print(" - df_payments")
    print(" - df_products")

except FileNotFoundError as e:
    print(f"Error: {e}")
    print("\nPlease make sure all 5 CSV files are uploaded to your Colab session.")
```

```python
print("--- Customers Info ---")
df_customers.info()
print("\n--- Orders Info ---")
df_orders.info()
print("\n--- Order Items Info ---")
df_order_items.info()
print("\n--- Payments Info ---")
df_payments.info()
print("\n--- Products Info ---")
df_products.info()
```

```python
df_customers.dropna(subset=['customer_id'], inplace=True)
```

```python
df_customers.drop_duplicates(subset=['customer_id'], keep='first', inplace=True)
```

```python
df_customers['name'].fillna('Unknown', inplace=True)
df_customers['email'].fillna('Unknown', inplace=True)
df_customers['address'].fillna('Unknown', inplace=True)
```

```python
df_customers['customer_id'] = df_customers['customer_id'].astype('Int64')
```

```python
df_customers['registration_date'] = pd.to_datetime(df_customers['registration_date'], errors='coerce')
```

```python
df_customers.info()
```

```python
print(df_customers.head())
```

```python
print(df_customers.tail())
```

```python
df_products.dropna(subset=['product_id'], inplace=True)
```

```python
df_products.drop_duplicates(subset=['product_id'], keep='first', inplace=True)
```

```python
df_products['product_name'].fillna('Unknown', inplace=True)
df_products['category'].fillna('Unknown', inplace=True)
```

```python
df_products['price'].fillna(0.0, inplace=True)
df_products['stock_quantity'].fillna(0, inplace=True)
```

```python
df_products['product_id'] = df_products['product_id'].astype('Int64')
df_products['stock_quantity'] = df_products['stock_quantity'].astype('Int64')
df_products['price'] = df_products['price'].astype(float)
```

```python
df_products.info()
```

```python
df_products.head()
```

```python
df_products.tail()
```

```python
df_orders.dropna(subset=['order_id', 'customer_id'], inplace=True)
```

```python
df_orders.drop_duplicates(subset=['order_id'], keep='first', inplace=True)
```

```python
df_orders['order_status'].fillna('Pendente', inplace=True)
```

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], errors='coerce')
```

```python
df_orders['order_date'].fillna("Unknown", inplace=True)
```

```python
df_orders['order_id'] = df_orders['order_id'].astype('Int64')
df_orders['customer_id'] = df_orders['customer_id'].astype('Int64')
```

```python
df_orders.info()
```

```python
df_orders.head()
```

```python
df_orders.tail()
```

```python
df_payments.dropna(subset=['payment_id', 'order_id'], inplace=True)
```

```python
df_payments.drop_duplicates(subset=['payment_id'], keep='first', inplace=True)
```

```python
df_payments['payment_method'].fillna('Unknown', inplace=True)
```

```python
df_payments['amount_paid'].fillna(0.0, inplace=True)
```

```python
df_payments['payment_date'] = pd.to_datetime(df_payments['payment_date'], errors='coerce')
```

```python
df_payments['payment_id'] = df_payments['payment_id'].astype('Int64')
df_payments['order_id'] = df_payments['order_id'].astype('Int64')
df_payments['amount_paid'] = df_payments['amount_paid'].astype(float)
```

```python
df_payments.info()
```

```python
df_payments.head()
```

```python
df_payments.tail()
```

```python
df_order_items.dropna(subset=['order_id', 'product_id'], inplace=True)
```

```python
df_order_items['quantity'].fillna(1, inplace=True)
```

```python
df_order_items['total_price'].fillna(0.0, inplace=True)
```

```python
df_order_items['order_id'] = df_order_items['order_id'].astype('Int64')
df_order_items['product_id'] = df_order_items['product_id'].astype('Int64')
df_order_items['quantity'] = df_order_items['quantity'].astype('Int64')
df_order_items['total_price'] = df_order_items['total_price'].astype(float)
```

```python
df_order_items.info()
```

```python
df_order_items.head()
```

```python
df_order_items.tail()
```

```python
cleaned_files = {
    "cleaned_customers.csv": df_customers,
    "cleaned_orders.csv": df_orders,
    "cleaned_order_items.csv": df_order_items,
    "cleaned_payments.csv": df_payments,
    "cleaned_products.csv": df_products
}
```

```python
for filename, df in cleaned_files.items():
    df.to_csv(filename, index=False)
    print(f"Successfully saved {filename}")

print("\nAll cleaned files have been saved!")
```
